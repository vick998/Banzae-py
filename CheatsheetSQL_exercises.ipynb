```python
import psycopg2 as pg2
#python-postgre interaction module
import pandas as pd
#dataframes and series
import re
#regex
import logins
conn = pg2.connect(database='dvd_rentals', user=logins.login,password=logins.master_pass)
cur = conn.cursor()
#connect to db and run sql commands to return data
cur.execute("SELECT * FROM film")
#store data locally
x = cur.fetchall()
cur.execute("SELECT COLUMN_NAME FROM INFORMATION_SCHEMA.COLUMNS WHERE TABLE_NAME = 'film' ORDER BY ORDINAL_POSITION")
x1 = list()
for i in cur.fetchall():
    x1.append(i[0])
s = pd.Series(x1)
df = pd.DataFrame.from_records(x, columns=s.values)
#creating df from db data
df
```

```python
df.length = df.length.apply(lambda x: str(x)+' minutes')
df.rental_rate = df.rental_rate.apply(lambda x: str(x)+' $')
df.last_update = df.last_update.apply(lambda x: str(x))
df.rental_duration = df.rental_duration.apply(lambda x: str(x) + ' months')
df.replacement_cost = df.replacement_cost.apply(lambda x: str(x) + ' $')
#int format data to str format
df
```

```python

s1 = df.last_update.str.split(' ')
s1l1 = list()
s1l2 = list()
for i in s1:
    s1l1.append(i[0])
    s1l2.append(i[1])
s1l1,s1l2
df['last_update_date'] = s1l1
df['last_update_time'] = s1l2
#split last_update column in 2 separate columns
df= df.drop(columns='last_update')
df
```

```python

df.fulltext = df.fulltext.str.split(' ')
df
#adding commas to fulltext column data
```

```python

df = df.drop(columns = ['language_id','release_year'])
#drop unneccessary columns and rows
df = df.drop(df[df.rating == 'R'].index)
#drop R rated films
df
```

```python
df = df.sort_values(by='film_id')
df = df.reset_index()
df = df.drop(columns='index')
df
#sorting the index values
```

```python
df.title = df.title.str.replace(' ','_')
df
#dropping whitespace in title column
```

```python
def sem(str0):
    if re.search('^A [AEIOU]',str0):
        str1 = 'An '+ str0[re.search('^A [AEIOU]',str0).end()-1:]
        return(str1)
    else:
        return(str0)
df.description = df.description.apply(sem)
df = df.drop(columns = ['special_features','fulltext'])
df
#editing desc, An for vowels, A for consonants
```
